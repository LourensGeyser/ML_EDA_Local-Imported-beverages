#### Based on the information above, combine the local and imported beverage as you deem appropriate [4]

### Building a models
#### Separate x and y data [2]

```python

x = df_beverages.iloc[:,:-1].values

y = df_beverages[df_beverages.columns[-1]]
```

#### Get the training and testing sets. Use 25% of the data for testing [5]

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
print('X_train')
print(X_train)
print('X_test')
print(X_test)
print('y_train')
print(y_train)
print('y_test')
print(y_test)
```

#### build a pipeline with standard scaler followed by logistic regression and fit the model [10]

```python
pipelr = Pipeline([('Scaler', StandardScaler()), ('Logistic regresion', LogisticRegression())])
pipelr.fit(X_train, y_train)
pipelr.score(X_test, y_test)
y_predlr = pipelr.predict(X_test)
```

 #### build a pipeline with standard scaler followed by Random Forest and fit the model [5]

```python
piperf = Pipeline([('Scaler', StandardScaler()), ('Random forest', RandomForestClassifier())])
piperf.fit(X_train, y_train)
piperf.score(X_test, y_test)
y_predrf = piperf.predict(X_test)
```

#### build a pipeline with standard scaler followed by  _Another Model of Your Choice_  and fit the model [5]

```python
pipenb = Pipeline([('Scaler', StandardScaler()), ('naive bayes', GaussianNB())])
pipenb.fit(X_train, y_train)
pipenb.score(X_test, y_test)
y_prednb = pipenb.predict(X_test)
```

#### Evaluate each model using:
1. A confusion matrix to see how the model's predictions align with the actual class labels.**[2 x 3]**
2. Accuracy to tell how many the model got right. **[1 x 3]**

```python
#logistic
print('confusion matrix\n')
print(pd.crosstab(y_test, y_predlr, rownames = ['Actual'], colnames = ['Predicted']))

print(accuracy_score(y_test,y_predlr))
print("\n")
print(classification_report(y_test, y_predlr))
```

```python
#random
print('confusion matrix\n')
print(pd.crosstab(y_test, y_predrf, rownames = ['Actual'], colnames = ['Predicted']))

print(accuracy_score(y_test,y_predrf))
print("\n")
print(classification_report(y_test, y_predrf))
```

```python
#naive
print('confusion matrix\n')
print(pd.crosstab(y_test, y_prednb, rownames = ['Actual'], colnames = ['Predicted']))

print(accuracy_score(y_test,y_prednb))
print("\n")
print(classification_report(y_test, y_prednb))
```
